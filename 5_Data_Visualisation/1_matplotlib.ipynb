Figures & subplot

```python
import matplotlib.pyplot as plt

fig = plt.figure()

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
```

```python
ax3 = fig.add_subplot(2, 2, 3) 
```

```python
from numpy.random import randn
ax3.plot(randn(50).cumsum(), 'k--')
```

```python
import numpy as np
_ = ax1.hist(randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * randn(30))
```

```python
fig
```

```python
fig, axes = plt.subplots(2, 3)
axes
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True) 
for i in range(2):
    for j in range(2):
        axes[i, j].hist(randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
```

Ticks, labels, legends

```python
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum())
```

```python
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')
```

```python
fig
```

```python
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum(), 'k', label='one')
ax.plot(randn(1000).cumsum(), 'k--', label='two')
ax.plot(randn(1000).cumsum(), 'k.', label='three')
ax.legend(loc='best')
```

Retour à Pandas

```python
# plot d'une Series
import pandas as pd

s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
s.plot()
```

```python
# plot d'un DataFrame
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0), columns=['A', 'B', 'C', 'D'], index=np.arange(0, 100, 10))
df.plot()
```

```python
# bar plot d'un DataFrame
fig, axes = plt.subplots(2, 1)
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))
data.plot(kind='bar', ax=axes[0], color='k', alpha=0.7)
data.plot(kind='barh', ax=axes[1], color='k', alpha=0.7)
```

```python
# bar plot d'un DataFrame
df = pd.DataFrame(np.random.rand(6, 4),
index=['one', 'two', 'three', 'four', 'five', 'six'],
columns=pd.Index(['A', 'B', 'C', 'D'], name='Genus'))

df.plot(kind='bar')
```

```python
comp1 = np.random.normal(0, 1, size=200) # N(0, 1)
comp2 = np.random.normal(10, 2, size=200) # N(10, 4)
values = pd.Series(np.concatenate([comp1, comp2]))
values.hist(bins=100, alpha=0.3, color='k', density= True)
values.plot.kde(style='k--')
```

```python
# scatter plot
# plot d'un DataFrame
df = pd.DataFrame(np.random.randn(50, 4), columns=['A', 'B', 'C', 'D'], index=np.arange(0, 100, 2))
df.plot.scatter(x="A", y="B", c="C", colormap = "Dark2")
```
