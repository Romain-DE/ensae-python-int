API

```python
import requests
url = 'https://api.github.com/events'
resp_github_events = requests.get(url)
json_resp = resp_github_events.json()
```

```python
json_resp
```

```python
commits_list = [resp["payload"]["commits"] for resp in json_resp if resp["payload"].get("commits")]
commits_list
```

```python
flat_commit_list = [
    commit
    for sublist in commits_list
    for commit in sublist
]
```

```python
import pandas as pd
events = pd.DataFrame(flat_commit_list)
events
```

```python
pd.set_option('display.max_colwidth', None)
events["author"]
```
