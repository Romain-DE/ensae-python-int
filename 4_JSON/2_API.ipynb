API

```python
import requests
url = 'https://api.github.com/events'
resp_github_events = requests.get(url)
json_resp = resp_github_events.json()
```

```python
json_resp
```

```python
existing_commits_list = [resp["payload"]["commits" ]for resp in json_resp if resp["payload"].get("commits")] # if len(resp["payload"]["commit"]) > 2
existing_commits_list
```

```python
flat_existing_commits_list = [
    commit
    for sublist in existing_commits_list
    for commit in sublist
]
```

```python
import pandas as pd
events = pd.DataFrame(flat_existing_commits_list)
events
```

```python
pd.set_option('display.max_colwidth', None)
events["author"]
```
