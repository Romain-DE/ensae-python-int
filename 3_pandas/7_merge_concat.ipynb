# Rappels

Jointure

```python
from pandas import DataFrame
df1 = DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'], 'data1': range(7)})
df1
```

```python
df2 = DataFrame({'key': ['a', 'b', 'd'], 'data2': range(3)})
df2
```

```python
# Par défaut, la jointure se fait sur la clé commune, avec le mode "left"
from pandas import merge
merge(df1, df2)
```

```python
# S'il n'y a pas de clé commune 
df3 = DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'], 'data1': range(7)})
df4 = DataFrame({'rkey': ['a', 'b', 'd'], 'data2': range(3)})

merge(df3, df4, left_on="lkey", right_on="rkey")
```

```python
# Différents modes de jointure
merge(df1, df2, how="right")
```

Concatenation

```python
from pandas import concat
concat([df1, df2], axis =1)["key"]
```

```python
# Par défaut, la concaténation est verticale
concat([df1, df2])
```

```python
concat([df1, df2], axis=1)
```

```python
# Concat sans contrainte d'Index
import numpy as np
df_1 = DataFrame(np.random.randn(3, 4), columns=['a', 'b', 'c', 'd'])
df_2 = DataFrame(np.random.randn(2, 3), columns=['b', 'd', 'a'])
```

```python
df_1
```

```python
df_2
```

```python
concat([df_1, df_2])
```

```python
concat([df_1, df_2], ignore_index=True)
```

```python
# Concat avec mode de jointure
from pandas import Series
s1 = Series([0, 1], index=['a', 'b'])
s2 = Series([2, 3, 4], index=['c', 'd', 'e']) 
s3 = Series([5, 6], index=['f', 'g'])
s4 = concat([s1 * 5, s3])
```

```python
s1
```

```python
s4
```

```python
df = concat([s1, s4])
df["b"]
```

```python
concat([s1, s4], axis=1, join='inner')
```

```python
concat([s1, s4], axis=1, join='outer')
```

Exercice

```python
# A partir des données du Titanic
# Créer une nouvelle colonne "prix moyen custom" qui correspond au prix moyen par catégorie "Pclass" et "Sex"
from pandas import read_csv
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df_titanic = read_csv(url).sample(100)

# Indice : pourquoi cet exercice se trouve-t-il dans cette catégorie ?
grouped_df = df_titanic.groupby(["Pclass", "Sex"])["Fare"].agg([("prix moyen custom", "mean")])
grouped_df
```

```python
merge(df_titanic, grouped_df, on = ["Pclass", "Sex"])
```

```python
# Bonus
# Une meilleure façon de faire : avec la fonction .transform()
df_titanic["prix moyen custom"] = df_titanic.groupby(["Pclass", "Sex"])["Fare"].transform("mean")
df_titanic.sort_values(["Pclass", "Sex"])
```
